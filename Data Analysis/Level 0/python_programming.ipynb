### Indexing and Slicing

Indexing allows you to access individual elements within sequence types like lists, tuples, and strings. Python uses zero-based indexing, meaning the first element is at index 0, the second at index 1, and so on.

- **Positive Indexing**: Starts from the beginning of the sequence (0, 1, 2, ...).


```python
from re import                                                                    # my_list       0:'apply' 1:'banana' ...
             |          |          |       |
             |          |          |       |
my_list = ['apple', 'banana', 'cherry', 'date']
        #     0         1         2       3
        #    -4       -3      -2         -1

print(my_list[0])   # Output: 'apple' (first element)
print(my_list[2])   # Output: 'cherry' (third element)
print(my_list[-1])  # Output: 'date' (last element
```

**Slicing** allows you to extract a portion (a "slice") of a sequence. It uses the syntax `[start:end:step]`, where `start` is the beginning index (inclusive), `end` is the ending index (exclusive), and `step` is the increment between elements.

```python
my_list = ['apple', 'banana', 'cherry', 'date', 'elderberry', 'fig']
       #     0         1        2        3        4              5
print(my_list[1:4:2])    # >> [1,4)
# Output: ['banana', 'cherry', 'date'] (elements from index 1 up to, but not including, 4)
print(my_list[:3])      # Output: ['apple', 'banana', 'cherry'] (elements from the beginning up to, but not including, 3)
print(my_list[3:])      # Output: ['date', 'elderberry', 'fig'] (elements from index 3 to the end)
print(my_list[::2])     # Output: ['apple', 'cherry', 'elderberry'] (every second element)

# my_list[start:end:step]
```

### Range

The `range` type represents an immutable sequence of numbers, commonly used for looping a specific number of times in `for` loops. It is memory-efficient as it generates numbers on demand rather than storing them all in memory.

- `range(stop)`: Generates numbers from 0 up to (but not including) `stop`.
- `range(start, stop)`: Generates numbers from `start` up to (but not including) `stop`.
- `range(start, stop, step)`: Generates numbers from `start` up to (but not including) `stop`, incrementing by `step`.

```python
for i in range(5):
    print(i, end=' ')
# Output: 0 1 2 3 4
print('\n')

for i in range(2, 8):
    print(i, end=' ')
# Output: 2 3 4 5 6 7
print('\n')

for i in range(1, 10, 2):
    print(i, end=' ')
# Output: 1 3 5 7 9
print('\n')

# To see the numbers as a list, you can convert it:
print(list(range(3)))
# Output: [0, 1, 2]
```

## 2. Conditional Statements (If-Else)

Conditional statements allow your program to make decisions and execute different code blocks based on whether certain conditions are met. The primary conditional statement in Python is `if-elif-else`.

```python
temperature = 25

if temperature > 30:
    print("It's a hot day!")
elif temperature > 20:
    print("It's a pleasant day.")
else:
    print("It's a bit chilly.")

score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
else:
    print("Grade: F")
```

## 3. Loops in Python

Loops allow you to run blocks of code multiple times.

### For Loop

Used to iterate over a sequence (like a list or a string).



```python
for i in range(5):
    print("Iteration:", i)
```

```python
fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
    print("I love" ,fruit)
```

```python
word = "Python"
for char in word:
    print(char, end=' ')
```



### While Loop

Repeats as long as a condition is `True`.



```python
x = 0
while x < 3:
    print("x is", x)
    x += 1
```



### Loop Control Statements

- `break` – Exit the loop early
- `continue` – Skip the rest of the loop and go to the next iteration



```python
for i in range(5):
    if i == 3:
        break
    print(i)
```

```python
for i in range(5):
    if i == 2:
        continue  # Skip printing 2
    print(i)
```

## 4. Functions in Python

Functions allow you to reuse blocks of code.

### Defining a Function



---

```python
def greet(name):
    print("Hello,", name)
```



### Calling a Function



```python
greet("Alice")  # Output: Hello, Alice
```



### Return Statement



```python
def add(a, b):
    return a + b

result = add(5, 3)
print(result)  # Output: 8
```



### Default Arguments



```python
def greet(name="friend"):
    print("Hi,", name)

greet()        # Output: Hi, friend
greet("Bob")   # Output: Hi, Bob
```



### Variable Scope

- Variables inside functions are **local**.
- Variables outside are **global** (unless shadowed).



```python
x = 5  # global

def change():
    x = 10  # local
    print("Inside function:", x)

change()
print("Outside function:", x)
```
