## Merging and Concatenating

### Explanation of Merging, Joining, and Concatenating

*   **Merging (`pd.merge()`)**: Used to combine two DataFrames based on common columns (keys). It's similar to SQL joins and offers various types (inner, outer, left, right).

*   **Concatenating (`pd.concat()`)**: Stacks DataFrames either vertically (row-wise) or horizontally (column-wise) along an axis. It's used to append or stack multiple DataFrames directly.

```python
# Create another DataFrame
dept = pd.DataFrame({'Name': ['Bob', 'Alice', 'Charlie', 'Alice'], 'Dept': ['HR', 'IT', 'Finance', 'CEO']})
print(dept)
print()
print(df_csv)
# Merge on 'Name'
pd.merge(df_csv, dept, on='Name')
```

```python

# Prepare an additional DataFrame to demonstrate outer, left, and right merges clearly
df_salaries = pd.DataFrame({'Name': ['Alice', 'Bob', 'Charlie', 'David'],
                            'Salary_2023': [52000, 61000, 72000, 55000]})

df_departments = pd.DataFrame({'Name': ['Alice', 'Bob', 'Eve'],
                                'Department': ['HR', 'IT', 'Marketing']})


print("DataFrames for merge examples:")
print("df_salaries:")
print(df_salaries)
print("\ndf_departments:")
print(df_departments)
print("\n")

print("Inner Merge (default):")
# Inner merge: Returns only the rows that have matching keys in both DataFrames.
merged_inner = pd.merge(df_salaries, df_departments, on='Name', how='inner')
print(merged_inner)
print("\n")

print("Outer Merge:")
# Outer merge: Returns all rows when there is a match in one of the DataFrames.
# If there is no match, the row is included with NaN values for the non-matching DataFrame's columns.
merged_outer = pd.merge(df_salaries, df_departments, on='Name', how='outer')
print(merged_outer)
print("\n")

print("Left Merge:")
# Left merge: Returns all rows from the left DataFrame, and the matching rows from the right DataFrame.
# If there is no match in the right DataFrame, NaN values are used for its columns.
merged_left = pd.merge(df_salaries, df_departments, on='Name', how='left')
print(merged_left)
print("\n")

print("Right Merge:")
# Right merge: Returns all rows from the right DataFrame, and the matching rows from the left DataFrame.
# If there is no match in the left DataFrame, NaN values are used for its columns.
merged_right = pd.merge(df_salaries, df_departments, on='Name', how='right')
print(merged_right)
```

## Grouping and Aggregating

### Explanation of Grouping and Aggregating

*   **Grouping**: The process of categorizing data into groups based on one or more keys (columns). This is often done using the `.groupby()` method.
*   **Aggregating**: After grouping, aggregation involves applying a function (like sum, mean, count, min, max, etc.) to each group to produce a single result for each group. This summarizes the data within each category.

```python
# Group by and calculate average salary
df_csv.groupby('Age').agg({'Salary': 'mean'})
```

## Applying Functions: `.apply()`, `.map()`, `.lambda`

### Explanation of `.apply()`, `.map()`, and `lambda`

*   **`.apply()`**: Used to apply a function along an axis of the DataFrame (rows or columns) or on Series values. It's very flexible and can be used for more complex operations.
*   **`.map()`**: Primarily used with a Series to substitute each value in the Series with another value, either by using a dictionary or a Series. It's optimized for one-to-one substitutions.
*   **`lambda`**: An anonymous, small, single-expression function. It's often used as an argument to higher-order functions like `apply()` or `map()` for concise, on-the-fly function definitions.

```python
# Using .apply with lambda
df_csv['Tax'] = df_csv['Salary'].apply(lambda x: x * 0.2)
df_csv
```

```python
# Using .apply with lambda
df_csv['Range'] = df_csv['Salary'].apply(lambda x: 'High' if x >= 60000 else 'Low')
df_csv
```

```python
status_map = {25: 'Young', 30: 'Adult', 35: 'Senior'}
df_csv['Status'] = df_csv['Age'].map(status_map)
df_csv
```

## Transforming Data: `.pivot()`, `.melt()`

### Explanation of Data Transformation Functions

*   **`.pivot()`**: Reshapes data based on column values. It creates a new table where unique values from one column become new columns, unique values from another column become the index, and values from a third column populate the new table.
*   **`.melt()`**: Unpivots a DataFrame from a 'wide' format to a 'long' format. It makes column headers into row values, useful for transforming data for visualization or analysis.


```python
monthly_sales_data = {
    'month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar'],
    'product': ['A', 'B', 'A', 'B', 'A', 'B'],
    'sales': [100, 150, 120, 180, 110, 160]
}
df_monthly_sales = pd.DataFrame(monthly_sales_data)
display(df_monthly_sales)
```

#### Pivoted Data: Transforming 'Long' to 'Wide' Format

```python
# Pivot the melted DataFrame back to a wide format
pivoted_monthly_sales = melted_monthly_sales.pivot(index='month', columns='product', values='value')
display(pivoted_monthly_sales)
```

#### Melted Data: Transforming 'Wide' to 'Long' Format

```python
# Reset the index to make 'month' a regular column again
unpivoted_df = pivoted_monthly_sales.reset_index()

# Melt the DataFrame to transform it back to a long format
# 'month' is the identifier variable
# The column headers ('A', 'B') become values in the 'product' column
# The cell values become values in the 'sales' column
result_df = unpivoted_df.melt(id_vars=['month'], var_name='product', value_name='sales')
result_df
```
