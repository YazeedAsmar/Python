# Pandas for Data Analysis
This notebook provides a beginner-friendly walkthrough of key pandas functionalities using simple examples. Topics include:
- Reading data from CSV, Excel, and JSON
- Selecting and filtering rows and columns
- Merging, concatenating datasets
- Grouping and aggregating data
- Applying functions with `.apply()`, `.map()`, and `.lambda`
- Transforming data with `.pivot()`, `.melt()`

```python
# Importing required libraries
import pandas as pd
import json
```

## Reading Data
We'll create small example files and read them using pandas.

```python
# Create example CSV file
csv_data = '''Name,Age,Salary
Alice,25,50000
Bob,30,60000
Charlie,35,70000'''
with open('example.csv', 'w') as f:
    f.write(csv_data)
# Read CSV
df_csv = pd.read_csv('example.csv')
df_csv
```

```python
# Create example Excel file
df_csv.to_excel('example.xlsx', index=False)
# Read Excel
df_excel = pd.read_excel('example.xlsx')
df_excel
```

```python
# Create example JSON file
json_data = df_csv.to_dict(orient='records')
with open('example.json', 'w') as f:
    json.dump(json_data, f)
# Read JSON
df_json = pd.read_json('example.json')
df_json
```

## Selecting and Filtering Rows and Columns

### Explanation of Selecting and Filtering Rows and Columns

*   **Selecting Columns**: In pandas, you can select one or more columns by using single or double square brackets with the column names. This returns a Series for a single column or a DataFrame for multiple columns.
*   **Filtering Rows**: You can filter rows based on a condition, which typically involves providing a boolean Series inside square brackets. Only rows where the condition is `True` will be returned.

```python
# Selecting columns
df_csv['Name']
```

```python
# Selecting multiple columns
df_csv[['Name', 'Age']]
```

```python
# Filtering rows where Age > 28
df_csv[df_csv['Age'] > 28]
```

```python
# Filtering rows with multiple conditions (Age > 28 AND Salary > 60000)
df_csv[(df_csv['Age'] > 28) & (df_csv['Salary'] > 60000)]
```

## Merging and Concatenating

### Explanation of Merging, Joining, and Concatenating

*   **Merging (`pd.merge()`)**: Used to combine two DataFrames based on common columns (keys). It's similar to SQL joins and offers various types (inner, outer, left, right).

*   **Concatenating (`pd.concat()`)**: Stacks DataFrames either vertically (row-wise) or horizontally (column-wise) along an axis. It's used to append or stack multiple DataFrames directly.

```python
# Create another DataFrame
dept = pd.DataFrame({'Name': ['Alice', 'Bob', 'Charlie'], 'Dept': ['HR', 'IT', 'Finance']})
# Merge on 'Name'
pd.merge(df_csv, dept, on='Name')
```

```python
# Concatenate two dataframes vertically
pd.concat([df_csv, df_csv])
```

## Grouping and Aggregating

### Explanation of Grouping and Aggregating

*   **Grouping**: The process of categorizing data into groups based on one or more keys (columns). This is often done using the `.groupby()` method.
*   **Aggregating**: After grouping, aggregation involves applying a function (like sum, mean, count, min, max, etc.) to each group to produce a single result for each group. This summarizes the data within each category.

```python
# Group by and calculate average salary
df_csv.groupby('Age').agg({'Salary': 'mean'})
```

## Applying Functions: `.apply()`, `.map()`, `.lambda`

### Explanation of `.apply()`, `.map()`, and `lambda`

*   **`.apply()`**: Used to apply a function along an axis of the DataFrame (rows or columns) or on Series values. It's very flexible and can be used for more complex operations.
*   **`.map()`**: Primarily used with a Series to substitute each value in the Series with another value, either by using a dictionary or a Series. It's optimized for one-to-one substitutions.
*   **`lambda`**: An anonymous, small, single-expression function. It's often used as an argument to higher-order functions like `apply()` or `map()` for concise, on-the-fly function definitions.

```python
# Using .apply with lambda
df_csv['Tax'] = df_csv['Salary'].apply(lambda x: x * 0.2)
df_csv
```

```python
# Using .apply with lambda
df_csv['Range'] = df_csv['Salary'].apply(lambda x: 'High' if x >= 60000 else 'Low')
df_csv
```

```python
status_map = {25: 'Young', 30: 'Adult', 35: 'Senior'}
df_csv['Status'] = df_csv['Age'].map(status_map)
df_csv
```

## Transforming Data: `.pivot()`, `.melt()`

### Explanation of Data Transformation Functions

*   **`.pivot()`**: Reshapes data based on column values. It creates a new table where unique values from one column become new columns, unique values from another column become the index, and values from a third column populate the new table.
*   **`.melt()`**: Unpivots a DataFrame from a 'wide' format to a 'long' format. It makes column headers into row values, useful for transforming data for visualization or analysis.


```python
# Melt example
melted = pd.melt(df_csv, id_vars=['Name'], value_vars=['Age', 'Salary'])
melted
```

```python
# Pivot table
df_pivot = melted.pivot(index='Name', columns='variable', values='value')
df_pivot
```
